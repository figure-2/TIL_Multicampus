```python
KAKAO_API_KEY = "ë‚´ API í‚¤"
KAKAO_WEB_SEARCH_URL = "https://dapi.kakao.com/v2/search/web" # ì£¼ì†Œ : ì¹´ì¹´ì˜¤-ë¬¸ì„œ-REST API-ê²€ìƒ‰/ê´‘ê³ /ê²°ì œ-Daum ê²€ìƒ‰-ì§€ì›í•˜ëŠ” ê¸°ëŠ¥-ì›¹ë¬¸ì„œ ê²€ìƒ‰í•˜ê¸°-URL
```

```python
header = {
  "Authorization": "KakaoAK {}".format(KAKAO_API_KEY)
}

header
```

```python
page = 50  # í˜ì´ì§€ ë²ˆí˜¸
size = 50  # í˜ì´ì§€ í¬ê¸°

# ê²€ìƒ‰ì–´ ì„¤ì •
query = "ìŠ¤ìš°íŒŒ2"

# ìš”ì²­ URLì— íŒŒë¼ë¯¸í„° ì¶”ê°€
params = {
    "query": query,
    "page": page,  # í˜ì´ì§€ ë²ˆí˜¸
    "size": size  # í˜ì´ì§€ í¬ê¸°
}
```

```python
import requests # íŒŒì´ì¬ì—ì„œ ìš”ì²­
response = requests.get(KAKAO_WEB_SEARCH_URL, headers=header, params=params)
response
```
